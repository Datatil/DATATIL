```python
# =======================
# Baseline DL : IMSK-only
# =======================
# - Lit le CSV "df_cleaned_no_nan_COLAB_IMSK.csv"
# - Colonne des chemins: 'IMSQ_path' (images masquées)
# - Quatre classes: ["COVID","Lung_Opacity","Normal","Viral Pneumonia"]
# - Split stratifié 80/10/10
# - Modèle simple: Conv(32,64,128) -> Dense(64,32) -> 4
# - Sans équilibrage/standardisation/SMOTE/weights/Z
# - Patience EarlyStopping: 8, epochs=100, batch=32
# - Affiche: accuracy, F1_macro, recall_macro, métriques par classe, courbes, confusion matrix

import os, re, random, gc, itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image
from google.colab import drive

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, f1_score, recall_score, precision_recall_fscore_support

import tensorflow as tf
from tensorflow.keras import layers, models

# -----------------------------
# 0) Paramètres "faciles à lire"
# -----------------------------
BASE      = "/content/drive/MyDrive/Radio_Covid19"
CSV       = f"{BASE}/df_cleaned_no_nan_COLAB_IMSK.csv"
PATH_COL  = "IMSQ_path"                      # colonne des chemins masqués
CLASS_ORDER = ["COVID","Lung_Opacity","Normal","Viral Pneumonia"]
IMG_SIZE  = (224, 224)
BATCH_SIZE= 32
EPOCHS    = 100
SEED      = 42

# -----------------------------
# 1) Monter Drive & charger CSV
# -----------------------------
drive.mount('/content/drive', force_remount=False)
assert os.path.exists(CSV), f"CSV introuvable : {CSV}"
df = pd.read_csv(CSV)

# Trouver la colonne label
label_col = None
for c in ["category","label","class"]:
    if c in df.columns:
        label_col = c; break
assert label_col is not None, "Aucune colonne d'étiquette ('category'/'label'/'class') trouvée."

# -----------------------------
# 2) Normaliser libellés & chemins
# -----------------------------
def norm_cls(c):
    c = str(c).strip()
    rep = {
        "Lung Opacity":"Lung_Opacity",
        "lung_opacity":"Lung_Opacity",
        "viral_pneumonia":"Viral Pneumonia",
        "Viral_Pneumonia":"Viral Pneumonia",
    }
    return rep.get(c, c)

def clean_path(p: str):
    if not isinstance(p, str): return ""
    q = p.strip().strip('"').strip("'").replace("\\","/")
    q = q.replace("My Drive","MyDrive")
    q = re.sub(r"/+", "/", q)
    if "Radio_Covid19" in q and not q.startswith("/content/drive"):
        q = "/content/drive/MyDrive/Radio_Covid19/" + q.split("Radio_Covid19/",1)[1]
    return q

df[label_col] = df[label_col].map(norm_cls)
df[PATH_COL]  = df[PATH_COL].apply(clean_path)

# Garder uniquement les fichiers existants (et dédoublonner par chemin)
df["__exists"] = df[PATH_COL].apply(os.path.exists)
df_ok = df[df["__exists"]].drop_duplicates(subset=[PATH_COL]).copy()

# Restreindre aux classes cibles, au cas où
df_ok = df_ok[df_ok[label_col].isin(CLASS_ORDER)].copy()
df_ok = df_ok.reset_index(drop=True)

# Comptage final
counts = (df_ok[label_col].value_counts()
          .reindex(CLASS_ORDER).fillna(0).astype(int))
print("\n=== Comptage IMSK par classe (fichiers réellement lisibles) ===")
for k,v in counts.items(): print(f"{k:16s}: {v:5d}")
print(f"TOTAL              : {int(counts.sum()):5d}")

# Encodage targets
cls_to_id = {c:i for i,c in enumerate(CLASS_ORDER)}
id_to_cls = {i:c for c,i in cls_to_id.items()}
df_ok["y"] = df_ok[label_col].map(cls_to_id).astype(int)

# --------------------------------------
# 3) Split stratifié 80/10/10 (train/val/test)
# --------------------------------------
df_train, df_tmp = train_test_split(
    df_ok, test_size=0.20, stratify=df_ok["y"], random_state=SEED
)
df_val, df_test = train_test_split(
    df_tmp, test_size=0.50, stratify=df_tmp["y"], random_state=SEED
)

def show_split_info():
    for name, d in [("TRAIN",df_train),("VAL",df_val),("TEST",df_test)]:
        c = d[label_col].value_counts().reindex(CLASS_ORDER).fillna(0).astype(int)
        print(f"\n{name} ({len(d)} images)")
        for k,v in c.items(): print(f"  {k:16s}: {v:5d}")
show_split_info()

# --------------------------------------
# 4) tf.data pipeline (lecture PNG/JPG -> 224x224 gris)
# --------------------------------------
AUTO = tf.data.AUTOTUNE
IMG_H, IMG_W = IMG_SIZE
N_CLASSES = len(CLASS_ORDER)

def decode_img(path):
    # tf.io.decode_image gère PNG/JPG, mais fixe le dtype et la shape
    img = tf.io.read_file(path)
    img = tf.io.decode_image(img, channels=1, expand_animations=False)
    img = tf.image.resize(img, [IMG_H, IMG_W], antialias=True)
    img = tf.cast(img, tf.float32) / 255.0
    return img

def build_ds(paths, y=None, training=False):
    x = tf.data.Dataset.from_tensor_slices(paths)
    x = x.map(decode_img, num_parallel_calls=AUTO)
    if y is not None:
        y = tf.one_hot(y, depth=N_CLASSES)
        y = tf.data.Dataset.from_tensor_slices(y)
        ds = tf.data.Dataset.zip((x, y))
    else:
        ds = x
    if training:
        ds = ds.shuffle(buffer_size=min(8192, len(paths)), seed=SEED, reshuffle_each_iteration=True)
    ds = ds.batch(BATCH_SIZE, drop_remainder=False)
    ds = ds.prefetch(AUTO)
    return ds

train_ds = build_ds(df_train[PATH_COL].values, df_train["y"].values, training=True)
val_ds   = build_ds(df_val[PATH_COL].values, df_val["y"].values, training=False)
test_ds  = build_ds(df_test[PATH_COL].values, df_test["y"].values, training=False)

# --------------------------------------
# 5) Modèle baseline (Conv 32-64-128 → Dense 64-32 → 4)
# --------------------------------------
def make_model():
    inputs = layers.Input(shape=(IMG_H, IMG_W, 1))
    x = layers.Conv2D(32, 3, padding="same")(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(64, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(128, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Flatten()(x)
    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dense(32, activation="relu")(x)
    outputs = layers.Dense(N_CLASSES, activation="softmax")(x)
    model = models.Model(inputs, outputs, name="IMSK_baseline_C32_64_128_D64_32_4")
    return model

model = make_model()
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
model.summary()

# --------------------------------------
# 6) Entraînement (EarlyStopping)
# --------------------------------------
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor="val_loss", patience=8, restore_best_weights=True
    )
]

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

# --------------------------------------
# 7) Évaluation sur TEST + métriques détaillées
# --------------------------------------
# Prédictions
y_proba = model.predict(test_ds, verbose=0)
y_pred  = np.argmax(y_proba, axis=1)

# Vérités terrain alignées à l'ordre de test_ds
y_true = df_test["y"].values

# Métriques globales
f1_macro     = f1_score(y_true, y_pred, average="macro")
recall_macro = recall_score(y_true, y_pred, average="macro")
acc          = (y_true == y_pred).mean()

print("\n=== METRIQUES GLOBALES (TEST) ===")
print(f"Accuracy      : {acc:.4f}")
print(f"F1_macro      : {f1_macro:.4f}")
print(f"Recall_macro  : {recall_macro:.4f}")

# Rapport par classe
report = classification_report(
    y_true, y_pred, target_names=CLASS_ORDER, output_dict=True, zero_division=0
)
rep_df = (pd.DataFrame(report)
          .rename_axis("Classe")
          .T.loc[CLASS_ORDER + ["accuracy","macro avg","weighted avg"]])

print("\n=== METRIQUES PAR CLASSE (TEST) ===")
display(rep_df.round(4))

# Sauvegarde optionnelle du tableau (décommente si besoin)
# out_csv = f"{BASE}/IMSK_baseline_metrics_test.csv"
# rep_df.round(6).to_csv(out_csv)
# print(f"Tableau métriques sauvegardé -> {out_csv}")

# --------------------------------------
# 8) Matrice de confusion
# --------------------------------------
cm = confusion_matrix(y_true, y_pred, labels=list(range(N_CLASSES)))
fig = plt.figure(figsize=(6,5))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Matrice de confusion — TEST")
plt.colorbar(fraction=0.046, pad=0.04)
tick_marks = np.arange(N_CLASSES)
plt.xticks(tick_marks, CLASS_ORDER, rotation=45, ha="right")
plt.yticks(tick_marks, CLASS_ORDER)
thresh = cm.max() / 2.
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, format(cm[i, j], 'd'),
             horizontalalignment="center",
             color="white" if cm[i, j] > thresh else "black")
plt.ylabel('Vrai label')
plt.xlabel('Label prédit')
plt.tight_layout()
plt.show()

# --------------------------------------
# 9) Courbes d'entraînement (loss & accuracy)
# --------------------------------------
hist = history.history
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(hist["loss"], label="train")
plt.plot(hist["val_loss"], label="val")
plt.title("Loss")
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.legend()

plt.subplot(1,2,2)
plt.plot(hist["accuracy"], label="train")
plt.plot(hist["val_accuracy"], label="val")
plt.title("Accuracy")
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend()
plt.tight_layout()
plt.show()

# --------------------------------------
# 10) Récapitulatif console succinct
# --------------------------------------
print("\n=== RÉSUMÉ BASELINE IMSK ===")
print(f"Images: train={len(df_train)}, val={len(df_val)}, test={len(df_test)}")
print(f"Acc (TEST)={acc:.4f} | F1_macro={f1_macro:.4f} | Recall_macro={recall_macro:.4f}")
```

```python
# =======================
# BASELINE IMSK — Recette 1
# =======================
# Modifs vs ta version précédente :
# - Architecture: C32 -> C64 -> C128 -> GAP -> Dense(4, softmax)
# - Loss: CategoricalCrossentropy(label_smoothing=0.05)
# - Callbacks: EarlyStopping (warm-up), ReduceLROnPlateau
# - Tout le reste conservé : epochs=100, batch=32, split stratifié, métriques, courbes.

import os, re, random, itertools, gc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image
from google.colab import drive

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, f1_score, recall_score

import tensorflow as tf
from tensorflow.keras import layers, models

# -----------------------------
# Paramètres généraux
# -----------------------------
BASE       = "/content/drive/MyDrive/Radio_Covid19"
CSV        = f"{BASE}/df_cleaned_no_nan_COLAB_IMSK.csv"
PATH_COL   = "IMSQ_path"
CLASS_ORDER= ["COVID","Lung_Opacity","Normal","Viral Pneumonia"]

IMG_SIZE   = (224, 224)
BATCH_SIZE = 32     # <= inchangé
EPOCHS     = 100    # <= inchangé
SEED       = 42
MIN_EPOCHS = 5      # warm-up avant que l’early-stopping ne s’active

# -----------------------------
# 1) Monter Drive & charger CSV
# -----------------------------
drive.mount('/content/drive', force_remount=False)
assert os.path.exists(CSV), f"CSV introuvable : {CSV}"
df = pd.read_csv(CSV)

# Trouver la colonne label
label_col = None
for c in ["category","label","class"]:
    if c in df.columns:
        label_col = c; break
assert label_col is not None, "Aucune colonne d'étiquette ('category'/'label'/'class') trouvée."

# -----------------------------
# 2) Normaliser libellés & chemins
# -----------------------------
def norm_cls(c):
    c = str(c).strip()
    rep = {
        "Lung Opacity":"Lung_Opacity",
        "lung_opacity":"Lung_Opacity",
        "viral_pneumonia":"Viral Pneumonia",
        "Viral_Pneumonia":"Viral Pneumonia",
    }
    return rep.get(c, c)

def clean_path(p: str):
    if not isinstance(p, str): return ""
    q = p.strip().strip('"').strip("'").replace("\\","/")
    q = q.replace("My Drive","MyDrive")
    q = re.sub(r"/+", "/", q)
    if "Radio_Covid19" in q and not q.startswith("/content/drive"):
        q = "/content/drive/MyDrive/Radio_Covid19/" + q.split("Radio_Covid19/",1)[1]
    return q

df[label_col] = df[label_col].map(norm_cls)
df[PATH_COL]  = df[PATH_COL].apply(clean_path)

# Garder uniquement les fichiers existants (dédoublonnage par chemin)
df["__exists"] = df[PATH_COL].apply(os.path.exists)
df_ok = df[df["__exists"]].drop_duplicates(subset=[PATH_COL]).copy()
df_ok = df_ok[df_ok[label_col].isin(CLASS_ORDER)].reset_index(drop=True)

# Comptage final
counts = (df_ok[label_col].value_counts()
          .reindex(CLASS_ORDER).fillna(0).astype(int))
print("\n=== Comptage IMSK par classe (lisibles) ===")
for k,v in counts.items(): print(f"{k:16s}: {v:5d}")
print(f"TOTAL              : {int(counts.sum()):5d}")

# Encodage
cls_to_id = {c:i for i,c in enumerate(CLASS_ORDER)}
id_to_cls = {i:c for c,i in cls_to_id.items()}
df_ok["y"] = df_ok[label_col].map(cls_to_id).astype(int)

# -----------------------------
# 3) Split stratifié 80/10/10
# -----------------------------
df_train, df_tmp = train_test_split(
    df_ok, test_size=0.20, stratify=df_ok["y"], random_state=SEED
)
df_val, df_test = train_test_split(
    df_tmp, test_size=0.50, stratify=df_tmp["y"], random_state=SEED
)

def show_split_info():
    for name, d in [("TRAIN",df_train),("VAL",df_val),("TEST",df_test)]:
        c = d[label_col].value_counts().reindex(CLASS_ORDER).fillna(0).astype(int)
        print(f"\n{name} ({len(d)} images)")
        for k,v in c.items(): print(f"  {k:16s}: {v:5d}")
show_split_info()

# -----------------------------
# 4) tf.data pipelines
# -----------------------------
AUTO = tf.data.AUTOTUNE
IMG_H, IMG_W = IMG_SIZE
N_CLASSES = len(CLASS_ORDER)

def decode_img(path):
    img = tf.io.read_file(path)
    img = tf.io.decode_image(img, channels=1, expand_animations=False)
    img = tf.image.resize(img, [IMG_H, IMG_W], antialias=True)
    img = tf.cast(img, tf.float32) / 255.0
    return img

def build_ds(paths, y=None, training=False):
    x = tf.data.Dataset.from_tensor_slices(paths).map(decode_img, num_parallel_calls=AUTO)
    if y is not None:
        y = tf.one_hot(y, depth=N_CLASSES)
        y = tf.data.Dataset.from_tensor_slices(y)
        ds = tf.data.Dataset.zip((x, y))
    else:
        ds = x
    if training:
        ds = ds.shuffle(buffer_size=min(8192, len(paths)), seed=SEED, reshuffle_each_iteration=True)
    return ds.batch(BATCH_SIZE).prefetch(AUTO)

train_ds = build_ds(df_train[PATH_COL].values, df_train["y"].values, training=True)
val_ds   = build_ds(df_val[PATH_COL].values,   df_val["y"].values,   training=False)
test_ds  = build_ds(df_test[PATH_COL].values,  df_test["y"].values,  training=False)

# -----------------------------
# 5) Modèle — Recette 1
#     C32 -> C64 -> C128 -> GAP -> Dense(4)
#     où chaque bloc: Conv(3x3) + BN + ReLU + MaxPool(2)
# -----------------------------
def make_model_recipe1(input_shape=(224,224,1), n_classes=4):
    In = layers.Input(shape=input_shape)

    x  = layers.Conv2D(32, 3, padding="same")(In)
    x  = layers.BatchNormalization()(x); x = layers.ReLU()(x)
    x  = layers.MaxPooling2D()(x)

    x  = layers.Conv2D(64, 3, padding="same")(x)
    x  = layers.BatchNormalization()(x); x = layers.ReLU()(x)
    x  = layers.MaxPooling2D()(x)

    x  = layers.Conv2D(128, 3, padding="same")(x)
    x  = layers.BatchNormalization()(x); x = layers.ReLU()(x)
    x  = layers.MaxPooling2D()(x)

    x  = layers.GlobalAveragePooling2D()(x)            # <-- GAP
    Out= layers.Dense(n_classes, activation="softmax")(x)
    return models.Model(In, Out, name="IMSK_recipe1")

model = make_model_recipe1(input_shape=(IMG_H, IMG_W, 1), n_classes=N_CLASSES)
model.summary()

# -----------------------------
# 6) Compilation + callbacks
# -----------------------------
loss = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.05)

callbacks = []
# EarlyStopping avec warm-up (débuter l'arrêt après MIN_EPOCHS)
try:
    es = tf.keras.callbacks.EarlyStopping(
        monitor="val_loss", patience=8, restore_best_weights=True,
        start_from_epoch=MIN_EPOCHS
    )
except TypeError:
    # fallback si version TF ancienne sans start_from_epoch
    es = tf.keras.callbacks.EarlyStopping(
        monitor="val_loss", patience=8, restore_best_weights=True
    )
callbacks.append(es)

callbacks.append(
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss", patience=3, factor=0.5, min_lr=1e-6, verbose=1
    )
)

model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
              loss=loss,
              metrics=["accuracy"])

# -----------------------------
# 7) Entraînement
# -----------------------------
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,                 # <= inchangé
    callbacks=callbacks,
    verbose=1
)

# -----------------------------
# 8) Évaluation & métriques
# -----------------------------
y_proba = model.predict(test_ds, verbose=0)
y_pred  = np.argmax(y_proba, axis=1)
y_true  = df_test["y"].values

acc          = (y_true == y_pred).mean()
f1_macro     = f1_score(y_true, y_pred, average="macro")
recall_macro = recall_score(y_true, y_pred, average="macro")

print("\n=== METRIQUES GLOBALES (TEST) ===")
print(f"Accuracy      : {acc:.4f}")
print(f"F1_macro      : {f1_macro:.4f}")
print(f"Recall_macro  : {recall_macro:.4f}")

report = classification_report(
    y_true, y_pred, target_names=CLASS_ORDER, output_dict=True, zero_division=0
)
rep_df = (pd.DataFrame(report)
          .rename_axis("Classe")
          .T.loc[CLASS_ORDER + ["accuracy","macro avg","weighted avg"]])
print("\n=== METRIQUES PAR CLASSE (TEST) ===")
display(rep_df.round(4))

# -----------------------------
# 9) Matrice de confusion
# -----------------------------
cm = confusion_matrix(y_true, y_pred, labels=list(range(N_CLASSES)))
fig = plt.figure(figsize=(6,5))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Matrice de confusion — TEST")
plt.colorbar(fraction=0.046, pad=0.04)
tick_marks = np.arange(N_CLASSES)
plt.xticks(tick_marks, CLASS_ORDER, rotation=45, ha="right")
plt.yticks(tick_marks, CLASS_ORDER)
thresh = cm.max() / 2.
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, format(cm[i, j], 'd'),
             horizontalalignment="center",
             color="white" if cm[i, j] > thresh else "black")
plt.ylabel('Vrai label'); plt.xlabel('Label prédit'); plt.tight_layout(); plt.show()

# -----------------------------
# 10) Courbes d'entraînement
# -----------------------------
hist = history.history
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(hist["loss"], label="train")
plt.plot(hist["val_loss"], label="val")
plt.title("Loss"); plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.legend()

plt.subplot(1,2,2)
plt.plot(hist["accuracy"], label="train")
plt.plot(hist["val_accuracy"], label="val")
plt.title("Accuracy"); plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend()
plt.tight_layout(); plt.show()

print("\n=== RÉSUMÉ BASELINE IMSK (Recette 1) ===")
print(f"Images: train={len(df_train)}, val={len(df_val)}, test={len(df_test)}")
print(f"Acc (TEST)={acc:.4f} | F1_macro={f1_macro:.4f} | Recall_macro={recall_macro:.4f}")
```
